**Limpeza e pré-processamento dos dados**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregar os arquivos CSV

df_agua = pd.read_csv('AbastAgua.csv', sep=";", encoding='utf-8')
df_lixo = pd.read_csv('DestLixo.csv', sep=";", encoding='utf-8')
df_esgoto = pd.read_csv('Esgoto.csv', sep=";", encoding='utf-8')
df_saneamento = pd.read_csv('Saneamento_Basico.csv', sep=";", encoding='utf-8')
df_fqescolar = pd.read_csv('Frequencia_escolar.csv', sep=";", encoding='utf-8')
df_medicos = pd.read_csv('medico_estado.csv', sep=";", encoding='utf-8')
df_censo = pd.read_csv('Censo_demografico_2022.csv', sep=";", encoding='utf-8')
```

```python
# Verificar as primeiras linhas de cada dataset

print("Abastecimento de água:")
print(df_agua.head().to_string())

print("\nColeta de lixo:")
print(df_lixo.head().to_string())

print("\nEsgoto:")
print(df_esgoto.head().to_string())

print("\nSaneamento básico:")
print(df_saneamento.head().to_string())

print("\nFrequencia escolar:")
print(df_fqescolar.head().to_string())

print("\nNúmero de médicos por estado:")
print(df_medicos.head().to_string())

print("\nCenso demográfico:")
print(df_censo.head().to_string())
```

```python
# Alterar nome de coluna para todos ficarem no mesmo padrão

df_agua.rename(columns={'Estados e  Capitais': 'Estados e Capitais'}, inplace=True)
df_saneamento.rename(columns={'Estados e  Capitais': 'Estados e Capitais'}, inplace=True)
df_fqescolar.rename(columns={'Estados e  Capitais': 'Estados e Capitais'}, inplace=True)
df_medicos.rename(columns={'Estados': 'Estados e Capitais'}, inplace=True)
```

```python
# Verificar tipos de dados

print("Tipagem Abastecimento de água:")
print(df_agua.dtypes)

print("\nTipagem Coleta de lixo:")
print(df_lixo.dtypes)

print("\nTipagem Esgoto:")
print(df_esgoto.dtypes)

print("\nTipagem Saneamento básico:")
print(df_saneamento.dtypes)

print("\nTipagem Frequencia escolar:")
print(df_fqescolar.dtypes)

print("\nTipagem Número de médicos por estado:")
print(df_medicos.dtypes)

print("\nTipagem Censo demográfico:")
print(df_censo.dtypes)
```

```python
# Verificar valores Nulos
print("Abastecimento de água:")
print(df_agua.isnull().sum())

print("\nColeta de lixo:")
print(df_lixo.isnull().sum())

print("\nEsgoto:")
print(df_esgoto.isnull().sum())

print("\nSaneamento básico:")
print(df_saneamento.isnull().sum())

print("\nFrequencia escolar:")
print(df_fqescolar.isnull().sum())

print("\nNúmero de médicos por estado:")
print(df_medicos.isnull().sum())

print("\nCenso demográfico:")
print(df_censo.isnull().sum())
```

```python
# Substituir valores nulos por 0

df_agua.fillna(0, inplace=True)
df_lixo.fillna(0, inplace=True)
df_esgoto.fillna(0, inplace=True)

print("Abastecimento de água:")
print(df_agua.isnull().sum())

print("\nColeta de lixo:")
print(df_lixo.isnull().sum())

print("\nEsgoto:")
print(df_esgoto.isnull().sum())
```

```python
# Remover colunas com dados não necessários

df_censo.drop(['Cód. UF', 'Cód. Município'], axis=1, inplace=True)

print("\nCenso demográfico:")
print(df_censo.head().to_string())
```

**Análises exploratórias**

```python
# Lista com os nomes do estados

estados_brasil = [
    "Acre", "Alagoas", "Amapá", "Amazonas", "Bahia", "Ceará", "Distrito Federal",
    "Espírito Santo", "Goiás", "Maranhão", "Mato Grosso", "Mato Grosso do Sul",
    "Minas Gerais", "Pará", "Paraíba", "Paraná", "Pernambuco", "Piauí",
    "Rio de Janeiro", "Rio Grande do Norte", "Rio Grande do Sul", "Rondônia",
    "Roraima", "Santa Catarina", "São Paulo", "Sergipe", "Tocantins"
]
```

1. Comparação entre os estados com maior e menor acesso a água tratada

```python
# Filtrar apenas os estados

df_agua_estados = df_agua[df_agua['Estados e Capitais'].isin(estados_brasil)]

# Ordenar os estados pelo percentual de acesso a rede de distribuição de água

df_agua_estados_ordenados = df_agua_estados.sort_values(by='Rede de distribuição', ascending=False)

# Mostrar os 5 estados com maior acesso a rede de distribução de água

print("Estados com maior acesso a rede de distribuição de água:")
print(df_agua_estados_ordenados[['Estados e Capitais', 'Região', 'Rede de distribuição']].head(5))

# Mostrar os 5 estados com menor acesso a rede de distribução de água

print("\nEstados com menor acesso a rede de distribuição de água:")
print(df_agua_estados_ordenados[['Estados e Capitais', 'Região', 'Rede de distribuição']].tail(5))
```

2. Comparação entre os estados com maior e menor coleta de lixo por serviço de limpeza

```python
# Filtrar apenas os estados

df_lixo_estados = df_lixo[df_lixo['Estados e Capitais'].isin(estados_brasil)]

# Ordenar os estados pelo percentual de coleta de lixo pelo serviço de limpeza

df_lixo_estados_ordenados = df_lixo_estados.sort_values(by='Coletado diretamente por serviço de limpeza', ascending=False)

# Mostrar os 5 estados com maior coleta de lixo por serviço de limpeza

print("Estados com maior coleta de lixo por serviço de limpeza:")
print(df_lixo_estados_ordenados[['Estados e Capitais', 'Região', 'Coletado diretamente por serviço de limpeza']].head(5).to_string())

# Mostrar os 5 estados com menor coleta de lixo por serviço de limpeza

print("\nEstados com menor coleta de lixo por serviço de limpeza:")
print(df_lixo_estados_ordenados[['Estados e Capitais', 'Região', 'Coletado diretamente por serviço de limpeza']].tail(5).to_string())
```

3. Comparação entre os estados com maior e menor cobertura a rede de esgoto

```python
# Filtrar apenas os estados

df_esgoto_estados = df_esgoto[df_esgoto['Estados e Capitais'].isin(estados_brasil)]

# Ordenar os estados pelo percentual de cobertura a rede de esgoto

df_esgoto_estados_ordenados = df_esgoto_estados.sort_values(by='Rede geral', ascending=False)

# Mostrar os 5 estados com maior cobertura a rede de esgoto

print("Estados com maior cobertura a rede de esgoto:")
print(df_esgoto_estados_ordenados[['Estados e Capitais', 'Região', 'Rede geral']].head(5).to_string())

# Mostrar os 5 estados com menor cobertura a rede de esgoto

print("\nEstados com menor cobertura a rede de esgoto:")
print(df_esgoto_estados_ordenados[['Estados e Capitais', 'Região', 'Rede geral']].tail(5).to_string())
```

4. Comparação entre os estados com maior e menor acesso ao saneamentro básico (acesso simultâneo aos três serviços, rede de aguá, esgoto e coleta de lixo)

```python
# Filtrar apenas os estados

df_saneamento_estados = df_saneamento[df_saneamento['Estados e Capitais'].isin(estados_brasil)]

# Ordenar os estados pelo percentual de acesso ao saneamentro básico

df_saneamento_estados_ordenados = df_saneamento_estados.sort_values(by='Acesso ao saneamento básico', ascending=False)

# Mostrar os 5 estados com maior acesso ao saneamento básico

print("Estados com maior acesso ao saneamento básico:")
print(df_saneamento_estados_ordenados[['Estados e Capitais', 'Região', 'Acesso ao saneamento básico']].head(5).to_string())

# Mostrar os 5 estados com menor acesso ao saneamento básico

print("\nEstados com menor acesso ao saneamento básico:")
print(df_saneamento_estados_ordenados[['Estados e Capitais', 'Região', 'Acesso ao saneamento básico']].tail(5).to_string())
```

5. Correlação entre os estados com maior e menor acesso a água tratada, coleta de lixo, esgoto e saneamento básico

```python
from functools import reduce

# Selecionar apenas as colunas relevantes de cada dataframe antes de mesclar

df_agua_infra = df_agua_estados_ordenados[['Estados e Capitais', 'Região', 'Rede de distribuição']]
df_lixo_infra = df_lixo_estados_ordenados[['Estados e Capitais', 'Coletado diretamente por serviço de limpeza']]
df_esgoto_infra = df_esgoto_estados_ordenados[['Estados e Capitais', 'Rede geral']]
df_saneamento_infra = df_saneamento_estados_ordenados[['Estados e Capitais', 'Acesso ao saneamento básico']]

dataframes_infra_filtrado = [df_agua_infra, df_lixo_infra, df_esgoto_infra, df_saneamento_infra]

# Merge todas as tabelas

df_merge_infraestrutura = reduce(lambda left, right: pd.merge(left, right, on="Estados e Capitais", how='outer'), dataframes_infra_filtrado)

print("Estados com maior infraestrutura de água, coleta de lixo, esgoto e saneamento básico:")
print(df_merge_infraestrutura.nlargest(5, 'Acesso ao saneamento básico').to_string())

print("\nEstados com menor infraestrutura de água, coleta de lixo, esgoto e saneamento básico:")
print(df_merge_infraestrutura.nsmallest(5, 'Acesso ao saneamento básico').to_string())
```

```python
# Calcular a correlação entre abastecimento de água, coleta de lixo, esgoto e saneamento básico

correlacao_infra = df_merge_infraestrutura[['Rede de distribuição', 'Coletado diretamente por serviço de limpeza', 'Rede geral', 'Acesso ao saneamento básico']].corr()

# Exibir a matriz de correlação

print("Matriz de Correlação entre Infraestrutura Urbana:")
print(correlacao_infra)
```

```python
# Gráfico da Correlação entre Infraestrutura Urbana

plt.figure(figsize=(8,6))
sns.heatmap(correlacao_infra, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlação entre Infraestrutura Urbana")
plt.show()
```

6. Comparação dos estados com a maior e menor frequência escolar do ensino fundamento ao ensino médio

```python
# Criar uma nova coluna com a média da frequência escolar

df_fqescolar['Média Frequência Escolar'] = df_fqescolar[['6 a 10 anos - anos iniciais do ensino fundamental',
                                                     '11 a 14 anos - anos finais do ensino fundamental',
                                                     '15 a 17 anos - ensino médio']].mean(axis=1)

# Filtrar apenas os estados

df_fqescolar_estados = df_fqescolar[df_fqescolar['Estados e Capitais'].isin(estados_brasil)]

# Ordenar os estados pelo percentual da média da frequência escolar

df_fqescolar_estados_ordenados = df_fqescolar_estados.sort_values(by='Média Frequência Escolar', ascending=False)

# Mostrar os 5 estados com a maior média da frequência escolar

print("Estados com a maior média da frequência escolar por estado:")
print(df_fqescolar_estados_ordenados[['Estados e Capitais', 'Região', 'Média Frequência Escolar']].head(5).to_string())

# Mostrar os 5 estados com a menor média da frequência escolar

print("\nEstados com a menor média da frequência escolar por estado:")
print(df_fqescolar_estados_ordenados[['Estados e Capitais', 'Região', 'Média Frequência Escolar']].tail(5).to_string())
```

7. Comparação dos estados com a maior e menor número de médicos por 10.000 habitantes em 2023

```python
# Ordenar os estados pelo número de médicos por a cada 10.000 habitantes em 2023

df_medicos_ordenados = df_medicos.sort_values(by='2023', ascending=False)

# Mostrar os 5 estados com o maior número de médicos por 10.000 habitantes em 2023

print("Estados com o maior número de médicos por 10.000 habitantes em 2023:")
print(df_medicos_ordenados[['Estados e Capitais', 'Região', '2023']].head(5).to_string())

# Mostrar os 5 estados com o menor número de médicos por 10.000 habitantes em 2023

print("\nEstados com o menor número de médicos por 10.000 habitantes em 2023:")
print(df_medicos_ordenados[['Estados e Capitais', 'Região', '2023']].tail(5).to_string())
```

8. Comparação dos estados com a maior e menor número total da população

```python
#Calcular a população total por estado

df_total_pop_estado = df_censo.groupby('UF')['Pop.Total'].sum().reset_index()

# Ordenar os estados pelo número total da população

df_total_pop_estado_ordenados = df_total_pop_estado.sort_values(by='Pop.Total', ascending=False)

# Mostrar os 5 estados com o maior número total da população

print("Estados com o maior número total da população:")
print(df_total_pop_estado_ordenados[['UF', 'Pop.Total']].head(5).to_string())

# Mostrar os 5 estados com o menor número total da população

print("\nEstados com o menor número total da população:")
print(df_total_pop_estado_ordenados[['UF', 'Pop.Total']].tail(5).to_string())
```

9. Correlação dos estados com maior e menor saneamento básico com a taxa de frequência escolar

```python
# Merge as tabelas de saneamento básico e de frequência escola

df_merge_saneamento_escola = pd.merge(df_saneamento_estados_ordenados, df_fqescolar_estados_ordenados, on='Estados e Capitais')

df_merge_saneamento_escola = df_merge_saneamento_escola[['Estados e Capitais', 'Região_x', 'Acesso ao saneamento básico', 'Média Frequência Escolar']]

print("Correlação entre os estados com maior saneamento e a frequência escolar:")
print(df_merge_saneamento_escola.head(5).to_string)

print("\nCorrelação entre os estados com menor saneamento e a frequência escolar:")
print(df_merge_saneamento_escola.tail(5).to_string)
```

```python
# Calcular correlação entre frequência escolar e saneamento básico

correlacao_escola_saneamento = df_merge_saneamento_escola["Média Frequência Escolar"].corr(df_merge_saneamento_escola["Acesso ao saneamento básico"])
print(f"Correlação entre frequência escolar e acesso ao saneamento básico: {correlacao_escola_saneamento:.2f}")
```

```python
# Gráfico para a visualizar a comparação entre saneamento e a frequencia ecolar

plt.figure(figsize=(8,6))
sns.scatterplot(data=df_merge_saneamento_escola, x="Acesso ao saneamento básico", y="Média Frequência Escolar", hue="Estados e Capitais", palette="coolwarm")
plt.xlabel("Porcentagem de Saneamento Básico (%)")
plt.ylabel("Taxa da Frequência Média Escolar (%)")
plt.title("Relação entre Saneamento Básico e Frequência Escolar")
plt.legend(title="Estado", loc="upper left", bbox_to_anchor=(1,1))
plt.grid(True)

plt.show()
```

10. Correlação dos estados com maior e menor saneamento básico com o número de médico por 10.000 habitantes

```python
# Merge as tabelas de saneamento básico e de número de médicos por a cada 10.000 habitantes

df_merge_saneamento_medico = pd.merge(df_saneamento_estados_ordenados, df_medicos_ordenados, on='Estados e Capitais')

df_merge_saneamento_medico = df_merge_saneamento_medico[['Estados e Capitais', 'Região_x', 'Acesso ao saneamento básico', '2023']]

print("Correlação entre os estados com maior saneamento e o número de médicos por a cada 10.000 habitantes:")
print(df_merge_saneamento_medico.head(5).to_string)

print("\nCorrelação entre os estados com menor saneamento e o número de médicos por a cada 10.000 habitantes:")
print(df_merge_saneamento_medico.tail(5).to_string)
```

```python
# Calcular correlação entre número de médicos por 10.000 habitantes e saneamento básico

correlacao_medico_saneamento = df_merge_saneamento_medico["2023"].corr(df_merge_saneamento_medico["Acesso ao saneamento básico"])
print(f"Correlação entre número de médicos por 10.000 habitantes e acesso ao saneamento básico: {correlacao_medico_saneamento:.2f}")
```

```python
# Gráfico para a visualizar a comparação entre saneamento e o número de médicos por a cada 10.000 habitantes

plt.figure(figsize=(8,6))
sns.scatterplot(data=df_merge_saneamento_medico, x="Acesso ao saneamento básico", y="2023", hue="Estados e Capitais", palette="coolwarm")
plt.xlabel("Porcentagem de Saneamento Básico (%)")
plt.ylabel("número de médicos por a cada 10.000 habitantes")
plt.title("Relação entre Saneamento Básico e número de médicos por a cada 10.000 habitantes")
plt.legend(title="Estado", loc="upper left", bbox_to_anchor=(1,1))
plt.grid(True)

plt.show()
```
